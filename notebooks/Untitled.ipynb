```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo11x.pt")
input_path = "crowd.mp4"
cap = cv2.VideoCapture(input_path)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

output_path = "crowd_yolo11.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame, classes=[0], imgsz=(1080, 1920), conf=0.2)[0]
    annotated = results.plot()
    out.write(annotated)

cap.release()
out.release()
```

```python
from ultralytics import RTDETR

model = RTDETR("rtdetr-l.pt")  # можно rtdetr-x.pt, rtdetr-m.pt

input_path = "crowd.mp4"
cap = cv2.VideoCapture(input_path)

fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

output_path = "crowd_rtdetr_conf.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(
        frame, conf=0.2, imgsz=(1080, 1920), agnostic_nms=True, classes=[0], iou=0.4
    )[0]
    annotated = results.plot()

    out.write(annotated)

cap.release()
out.release()
```

```python
input_path = "crowd_yolo11.mp4"
output_path = "crowd_yolo11_resized.mp4"
cap = cv2.VideoCapture(input_path)

new_width = 480
new_height = 270

fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

out = cv2.VideoWriter(output_path, fourcc, fps, (new_width, new_height))

while True:
    ret, frame = cap.read()
    if not ret:
        break
    resized_frame = cv2.resize(frame, (new_width, new_height))
    out.write(resized_frame)

cap.release()
out.release()
```
